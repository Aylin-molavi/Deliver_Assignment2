```python
def reverse_complement(dna):
    complement = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return ''.join(complement[base] for base in reversed(dna))

def reversepalindromes(dna):  #between 4 and 12 in a DNA string
    reverse_palindromes = []
    
    # substrings of lengths between 4 and 12
    for length in range(4, 13):
        for i in range(len(dna) - length + 1):
            substring = dna[i:i + length]
            if substring == reverse_complement(substring):
                reverse_palindromes.append((i + 1, length))  # Store 1-based index
    
    return reverse_palindromes

def read_fasta(filename):
    with open(filename, 'r') as file:
        dna = ""
        for line in file:
            if not line.startswith(">"):
                dna += line.strip()  # Concatenate the sequences
    return dna

filename = "Datasets/rosalind_revp.txt" 
dna_string = read_fasta(filename)
result = reversepalindromes(dna_string)

for start, length in result:
    print(start, length)
```
